```python
from sympy.parsing.sympy_parser import parse_expr
from sympy import Matrix, symbols, expand, latex
import numpy as np
import matplotlib.pyplot as plt
```

```python
f = parse_expr('2*sin(6 * phi)')
var = list(f.free_symbols)[0]

r = np.arange(0, 2*pi, 0.01)
theta = [f.subs(var, r).evalf() for r in r]

plt.figure(figsize=(10,10))
ax = plt.subplot(111, projection='polar')
ax.plot(r, theta, c='purple', lw=4)
ax.grid(True)

ax.set_title("Polat plot of ${}$".format(latex(f)), va='bottom')
plt.show()
```
