```python
t = symbols('t')
u = parse_expr('Matrix([t*cos(3*t), t*sin(2*t), t])')
a = 16
lim = [0, a]
ts = np.linspace(lim[0], lim[1], a**2)
Y = np.array([u.subs('t', t).evalf() for t in ts]).astype('float')
xs = [x[0] for x in Y]
ys = [x[1] for x in Y]
zs = [x[2] for x in Y]
fig = plt.figure(figsize=(10,10))
#plt.rcParams['savefig.dpi'] = 600
ax = fig.add_subplot(111, projection='3d')
plot(xs, ys, zs, c="purple")
plt.show()
```

# 1 var extremas

```python
x = symbols('x')
expr = parse_expr('-x**3 + 10*x**2 - x')
xs = np.linspace(-2, 10, 20)
ys = np.array([expr.subs('x', x).evalf() for x in xs]).astype('float')
plot(xs, ys)
```

```python
expr.diff('x')
```

```python
expr_dx = expr.diff(x)
xs = np.linspace(-2, 10, 20)
ys = np.array([expr_dx.subs('x', x).evalf() for x in xs]).astype('float')
plot(xs, ys)
axhline(0, c='black', lw=1, ls='dashed')
cp = solve(expr_dx, 'x')
plot(cp, 'o')
```
